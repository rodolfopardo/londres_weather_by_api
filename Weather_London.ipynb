### Tiempo en Londres mediante petición por API 

#### Api: https://www.metaweather.com/api/

```python
#Importando librerías a utilizar 

import requests
import pandas as pd 
```

```python
#Metodo get para llamar

weather_london = requests.get("https://www.metaweather.com/api/location/44418/2019/7/14/")
```

```python
#Mostramos la conexion satisfactoria
weather_london
```

```python
#Leemos la data 

weather_json = weather_london.json()
```

```python
#Imprimimos 
weather_json
```

```python
#Realizamos un dataframe 

london_df = pd.DataFrame(weather_json)
```

```python
#Imprimimos DataFrame

london_df
```

```python
#Las columnas con sus tipos de datos 

london_df.dtypes
```

```python
#Algunos resultados estadísticos 

london_df.describe()
```

```python
#Tratamos de identificar valores nulos 

london_df.isnull()
```
